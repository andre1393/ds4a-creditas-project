```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
converters = {}
replacer = lambda x: float(x.replace(',','.')) if (x != 'X' and x != '') else None

for i in range(1, 13):
    converters[f'BASICO_V{str(i).zfill(3)}'] = replacer
for i in range(1, 15):
    converters[f'DOMICILIO_RENDA_V{str(i).zfill(3)}'] = replacer
for i in range(1, 133):
    converters[f'PESSOA_RENDA_V{str(i).zfill(3)}'] = replacer

dcctv = pd.read_csv('../data/dados_censitarios_consolidados_todas_variaveis.csv', index_col=0, converters=converters)
```

```python
grp = dcctv.groupby('Nome_da_UF ').mean()
```

```python
grp[[f'BASICO_V{str(i).zfill(3)}' for i in range(1, 13)]].plot(kind='bar', figsize=(15,8), grid=True)
```

```python
grp[[f'DOMICILIO_RENDA_V{str(i).zfill(3)}' for i in range(1, 15)]].plot(kind='bar', figsize=(15,8), grid=True, logy=True)
```

```python
registry = pd.read_csv('../data/registry.csv')
registry = registry.dropna(subset=['coordinates'])
```

```python
registry.head()
```

```python
registry.info()
```

```python
registry.sort_values('coordinates')
```

```python
registry['long'] = registry['coordinates'].apply(lambda x: float(x.split(',')[0][1:]))
registry['lat'] = registry['coordinates'].apply(lambda x: float(x.split(',')[1][:-1]))
```

```python
registry = registry[(registry['lat'] < 6) & (registry['lat'] > -34) & (registry['long'] > -74) & (registry['long'] < -34)]
```

```python
c = registry['analyst_decision'].apply(lambda x: 'blue' if x=='A' else 'red' if x=='R' else 'green')
registry.plot(kind='scatter', x='long', y='lat', c=c, grid=True, figsize=(6,6))
```

```python
registry\
.groupby('analyst_decision')\
.count()\
.reset_index()\
.plot(kind='bar', x='analyst_decision', y='model_decision', grid=True, figsize=(15,5), legend=False)\
```

```python
registry\
.groupby(['analyst_decision', 'model_decision'])\
.count()\
.reset_index()\
.pivot(index='model_decision', columns='analyst_decision', values='zip_code')\
.plot(kind='bar', figsize=(15,9), stacked=True, grid=True)
```
