# Registry - Analyst review history

```python
def clean_str(x):
    return unidecode.unidecode(x.title().strip().replace('  ', ' '))
```

```python
registry = pd.read_csv('../data/registry.csv')
registry = registry.dropna(subset=['coordinates'])
```

```python
registry.head()
```

```python
registry.info()
```

### Approved/Rejected ratio

```python
registry\
.groupby('analyst_decision')\
.count()\
.reset_index()\
.plot(kind='bar', x='analyst_decision', y='model_decision', grid=True, figsize=(15,5), legend=False)\
```

```python
registry\
.groupby(['analyst_decision', 'model_decision'])\
.count()\
.reset_index()\
.pivot(index='model_decision', columns='analyst_decision', values='zip_code')\
.plot(kind='bar', figsize=(15,9), stacked=True, grid=True)
```

### State

```python
registry['state'] = registry['concat'].str[-2:]
(100 * registry['state'].value_counts()/len(registry)).plot(kind='bar', grid=True, figsize=(15,5), title='percent of leads by state')
```

```python
registry\
.groupby(['state', 'analyst_decision'])\
.count()['concat']\
.reset_index()\
.pivot_table(index='state', columns='analyst_decision', values='concat')\
.plot(kind='bar', grid=True, figsize=(15,8), log=True, title='Approved/Reject ratio by state (log scale)')
```

### City

```python
import unidecode
registry['city'] = registry['concat'].apply(lambda x: clean_str(x.split(',')[-2]))
(100 * registry['city'].value_counts()/len(registry))[:50].plot(kind='bar', figsize=(15,8), grid=True, title='Top 50 cities in percent')
```

```python
sp = registry[registry['state'] == 'SP']
a = ((sp['city'] == 'Sao Paulo').value_counts()/len(sp))
a.index = ['Others', 'São Paulo']
a.plot(kind='bar', figsize=(15, 8), grid=True, title='Lead in SP in percet')
```

```python
pt = registry[registry['state'] == 'SP']\
.groupby(['city', 'analyst_decision'])\
.count()\
.reset_index()\
.pivot_table(index='city', columns='analyst_decision', values='model_decision')\

pt['total'] = pt['A'] + pt['R']
pt.sort_values(by='total', ascending=False).iloc[:40, :2]\
.plot(kind='bar', grid=True, figsize=(15,8), log=True, title='Approved/Reject ratio by city (log scale)')
```

### lat/long

```python
# extracting lat/long
registry['long'] = registry['coordinates'].apply(lambda x: float(x.split(',')[0][1:]))
registry['lat'] = registry['coordinates'].apply(lambda x: float(x.split(',')[1][:-1]))
```

```python
# removing lat/long that is not in Brazil
registry = registry[(registry['lat'] < 6) & (registry['lat'] > -34) & (registry['long'] > -74) & (registry['long'] < -34)]
```

```python
c = registry['analyst_decision'].apply(lambda x: 'blue' if x=='A' else 'red' if x=='R' else 'green')
registry.plot(kind='scatter', x='long', y='lat', c=c, grid=True, figsize=(6,6))
```

### Neighborhood

```python
registry['neighborhood'] = registry['concat'].apply(lambda x: clean_str(x.split(',')[1]))
```

```python
pt = registry\
.groupby(['neighborhood', 'analyst_decision'])\
.count()\
.pivot_table(index='neighborhood', columns='analyst_decision', values='concat')

pt['total'] = pt['R'] + pt['A']
pt.sort_values(by='total', ascending=False)\
.iloc[:30,:2]\
.plot(kind='bar', figsize=(15,8), grid=True, title='Neighborhoods in São Paulo')
```
